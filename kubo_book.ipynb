```r
y <- 0:20
prob <- dpois(y, lambda = 3.5)
plot(y, prob, xlim=c(0,20), ylim=c(0,0.21), ann=F, type = "b", pch = 1, lty = 2)
par(new=T)
prob <- dpois(y, lambda = 7.7)
plot(y, prob, xlim=c(0,20), ylim=c(0,0.21), ann=F, type = "b", pch = 5, lty = 2)
par(new=T) 
prob <- dpois(y, lambda = 15.1)
plot(y, prob, xlim=c(0,20), ylim=c(0,0.21), type = "b", pch = 2, lty = 2)
legend("topright", legend = c("3.5","7.7","15.1"), pch = c(1,5,2), lty = c(0,0,0), title="lambda")
```

```r
logL <- function(m) sum(dpois(data, m, log = TRUE))

plot.poisson <- function(lambda) {
    y <- 0:9
    prob <- dpois(y, lambda = lambda)

    hist(data, breaks = seq(-0.5, 9.5, 1), ylim = c(0, 15),
         main = "", xlab = "", ylab = "")
    points(y, prob * 50)
    lines(y,  prob * 50, lty = 2)

    title(sprintf("                    lambda= %.1f\n                    logL= %.1f", lambda, logL(lambda)), line = -2)
}

layout(matrix(1:9, byrow = T, ncol = 3))
junk <- sapply(seq(2, 5.2, 0.4), plot.poisson)
```
